```python
import multiprocess
import time
```

```python
def f(x):
    ans = 0
    data,y = x
    for i in data:
        ans += i**i
    return ans,y
 
```

```python
x = [i for i in range(10000)]
```

```python
start = time.time()
ans = []
for i in range(3):
    ans.append(f(x))
end = time.time()
print(len(ans))
print(end-start)
```

```python
if __name__ == '__main__':
    # 依 CPU 數量建立 child process
    pool = multiprocess.Pool(processes=3)
    start = time.time()
    res = pool.map(f,[(x,1),(x,1),(x,1)])
    end = time.time()
    print(len(res))
    print(end-start)
```
