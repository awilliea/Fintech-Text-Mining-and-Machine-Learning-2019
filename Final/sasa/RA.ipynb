```python
import numpy as np
import pandas as pd
from pyquery import PyQuery as pq
import time
import datetime
import requests
import urllib.request
import os
import calendar
import pickle
from selenium import webdriver
```

```python
file = pd.read_csv('chien.csv')
file.head()
```

```python
#find all tickers
ticker = []
for n in list(set(file['Ticker'])):
    ticker.append(n)
ticker
```

```python
#find all search address
#def get_the_file_address
#得到大網站
for tic in ticker[:2]:
    browser = webdriver.Chrome("./chromedriver.exe") 
    browser.get('https://www.sec.gov/edgar/searchedgar/legacy/companysearch.html')
    browser.find_element_by_xpath("//input[@title='CIK']").send_keys(tic)
    browser.find_element_by_xpath("//input[@value='Find Companies']  ").click()
    browser.find_element_by_xpath("//input[@name='type']").send_keys("DEF")
    browser.find_element_by_xpath("//input[@value='Search']  ").click()
    time.sleep(1)
    c_url = browser.current_url
    time.sleep(1)
    
```
