```python
Corr = {}
Keys = CL
for n in range(len(Keys)):
    data = file[['Date',Keys[n]]]
    d1 = datetime.strptime(data['Date'][1], '%Y-%m-%d')
    if d1 <=msci['Date'][1]:            #設立早於1996-01-02
        loc = data[data['Date']=='1996-01-02'].index.item()
        data = data[loc:]
        try:
            corr = data[Keys[n]].corr(msci['Returns'][1:], method='pearson')
        except:
            corr = data[Keys[n]].replace("C",0).corr(msci['Returns'][1:], method='pearson')
        Corr[Keys[n]] = corr
    else:                               #設立晚於1996-01-02
        loc = msci[msci['Date']==data['Date']].index.item()
        msci = msci[loc:]
        corr = data[Keys[n]].corr(msci['Returns'][1:], method='pearson')
        Corr[Keys[n]] = corr
print(Corr)
```
